```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
   missing = np.mean(df[col].isnull())
   print('{} - {}%'.format(col, missing))
```

```python
df.dtypes
```

```python
#Replacing all NAN values with 0

df.fillna(value=0, inplace=True)
```

```python
#chanding datatypes

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['company'] = df['company'].astype('str')
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.drop_duplicates()
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
#plotting budget vs gross using sns

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"#43a2ca"}, line_kws={"color":"#31a354"})

#the line shows corelation, it is going which means a positive corelation 
```

```python
#Looking for corelation
df.corr(numeric_only = True) # different corelations - pearson(default), kendall, spearman 
```

```python
# hight corelation b/w budget and gross
```

```python
correlation_matrix = df.corr(numeric_only = True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# company
# company is not a numeric value

df_toNumeric = df
for col_name in df_toNumeric.columns:
    if(df_toNumeric[col_name].dtype == 'object'):
        df_toNumeric[col_name] = df_toNumeric[col_name].astype('category')
        df_toNumeric[col_name] = df_toNumeric[col_name].cat.codes
        
df_toNumeric        
```

```python
correlation_matrix = df_toNumeric.corr(numeric_only = True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_toNumeric.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_cpairs = corr_pairs.sort_values()
sorted_cpairs
```

```python
high_corr = sorted_cpairs[(sorted_cpairs)>0.5]
high_corr
```
